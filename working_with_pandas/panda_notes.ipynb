Row selection

`iloc` is used to select rows/columns or rows and column using index slicing.
This is very useful especially when your data bset do not have labels(that is, row names and column names)

````python
# selecting some rows
df.iloc[0:7] # starts from the zero index and stops at the 6th element
````

````python
# combination of row and column selection
df.iloc[0:5, 0:4]#the first slice picks the rows (0-4) and the second slice prints column(0-3)
````



**Difference between `loc` and `iloc`**
- `loc` is label based(prints the first and last element specified) and `iloc` is index based (Uses zero index, prints the first element but doesn't print the last element specified)
- `loc` doesn't allow comnbination of rows and columns, `iloc` allows.


````python
# using iloc()

df.loc[7:] #prints the seventh to the last element
df.loc[:7] #prints the 0th to the last element

df.loc[::]
````

Conditional Filtering

````python
# filter rows where Gender is Male
filtered_male = df[df["Gender"] == "Male"]
print("Rows where Gender is Male")
filtered_male

# filter rows where Course Track is AI
filtered_course = df[df["Course Track"] == "AI"]
replaced_course = filtered_course.replace("AI", "Cybersecurity")
print("Rows where Course Track is AI")
replaced_course


# filter rows where PC-Make is Macbook or Dell
pc_make = df[(df["PC-Make"]== "MACBOOK") & (df["PC - OS"]== "Mac OS")]
print("rows where PC-Make is  or Dell")
pc_make


# Filter rows where City is either 'Lagos' or 'Abeokutaa'
cities = ['Lagos', 'Abeokuta']
city_filtered = df[df['City'].isin(cities)]
print("Rows where City is either 'Lagos' or 'Abeokuta':")
city_filtered
````
